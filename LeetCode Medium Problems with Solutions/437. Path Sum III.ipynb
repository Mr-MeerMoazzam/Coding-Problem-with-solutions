#Problem Description
Given the root of a binary tree and an integer targetSum, return the number of paths where the sum of the values along the path equals targetSum.

The path does not need to start or end at the root or a leaf, but it must go downwards (i.e., traveling only from parent nodes to child nodes).

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
     def sum_containing_root(self, root, target_sum, curr_sum):
        if root == None: return 0
        curr_sum = curr_sum + root.val
        left_sub_tree = self.sum_containing_root(root.left, target_sum, curr_sum)
        right_sub_tree = self.sum_containing_root(root.right, target_sum, curr_sum)
        count = 1 if curr_sum == target_sum else 0
        return count + left_sub_tree + right_sub_tree
    
     def pathSum(self, root, targetSum):
        if root == None: return 0
        
        count_left = self.pathSum(root.left, targetSum)
        count_right = self.pathSum(root.right, targetSum)
        
        return self.sum_containing_root(root, targetSum, 0) + count_left + count_right
```
