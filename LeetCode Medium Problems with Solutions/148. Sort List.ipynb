#Problem Description
Given the head of a linked list, return the list after sorting it in ascending order.

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def sortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        #if there is just one or no element in the linked list no need for any action 
        if head is None or head.next is None:
            return head
        
        #take all the values from linked list and put into a list
        pointer = head
        nums = []
        while pointer is not None:
            nums.append(pointer.val)
            pointer = pointer.next
        
        #sort the list
        nums.sort()
        #then create a new linked list and put all the sorted list values into the newly created linked list
        currentNode = ListNode(nums[0], None)
        head = currentNode
        for x in nums[1:-1]:
            currentNode.next = ListNode(x, None)
            currentNode = currentNode.next
        currentNode.next = ListNode(nums[-1], None)
        return head
```
