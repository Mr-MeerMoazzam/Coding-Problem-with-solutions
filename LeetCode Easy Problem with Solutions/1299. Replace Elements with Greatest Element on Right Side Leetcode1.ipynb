#Problem Description
Given an array arr, replace every element in that array with the greatest element among the elements to its right, and replace the last element with -1.

After doing so, return the array.

```python
class Solution(object):
    def replaceElements(self, arr):
        """
        :type arr: List[int]
        :rtype: List[int]
        """
        len_arr=len(arr)
        #if length of the list is 1 then it simply replace it with -1
        if len_arr==1:
            arr[0]=-1
            return arr
        #else it will iterate over the list from second element and replace the values with a right value as explained in problem discription
        else:
            for i in range(1,len_arr):
                arr[i-1]=max(arr[i:])
        arr[-1]=-1        
        return arr        
        
```
