#Problem Description
Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

```python
class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        #simply convert the nums into set(duplicates will be removed if any and length will be less than original) and compare the lengths before and after conversion
        return False if len(nums)==len(list(set(nums))) else True
        
```
