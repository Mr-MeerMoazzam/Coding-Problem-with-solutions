#Problem Description
Given an integer numRows, return the first numRows of Pascal's triangle.

In Pascal's triangle, each number is the sum of the two numbers directly above 
Example 1:

Input: numRows = 5
Output: [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]
Example 2:

Input: numRows = 1
Output: [[1]]

```python
class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        arr=[]
        for i in range(0,numRows):
            inner=[]
            for j in range(0,i+1):
                #inner list first index will be filled with 1
                if j==0:
                    inner.append(1)
                #inner list last index will aslo be filled with 1
                elif j==i:
                    inner.append(1)
                #inner list indexes other then 1st and last will be filled with summation of upper two index of the lists
                else:
                    inner.append(arr[i-1][j-1]+arr[i-1][j])
            arr.append(inner)
        return arr
```
