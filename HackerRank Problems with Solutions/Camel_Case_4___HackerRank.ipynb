<b>Sample Output</b><br>

large software book<br>

whiteSheetOfPaper()<br>

picture frame<br>

<b>Explanation</b><br>

Use Scanner to read in all information as if it were coming from the keyboard.<br>

Print all information to the console using standard output (System.out.print() or System.out.println()).<br>

Outputs must be exact (exact spaces and casing).<br>

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
def split_into_words(s):
  words=[]
  ind=0
  for i, c in enumerate(s):
      if c.isupper():
        words.append(s[ind:i].lower())
        ind=i
  words.append(s[ind:].lower())
  return words

while True:
    try:
        i = input().rstrip()  
        op_name=i.split(';')[0]
        obj_type=i.split(';')[1]
        obj_name=i.split(';')[2]
        if op_name=='S':
            if obj_type=='V' or obj_type=='C':
                words=split_into_words(obj_name)
                words = list(filter(None, words))
                print(" ".join(words))
            else:
                words=split_into_words(obj_name[:-2])
                words = list(filter(None, words))
                print(" ".join(words))
        else:
            if obj_type=='V':
                combine_words=[word if i==0 else word.capitalize() for i,word in enumerate(obj_name.split())] 
                print("".join(combine_words))
            elif obj_type=='C':
                combine_words=[word.capitalize() for word in obj_name.split()] 
                print("".join(combine_words))
            elif obj_type=='M':
                combine_words=[word if i==0 else word.capitalize() for i,word in enumerate(obj_name.split())]
                combine_words.append('()')
                print("".join(combine_words))
    except EOFError:
        break
```
