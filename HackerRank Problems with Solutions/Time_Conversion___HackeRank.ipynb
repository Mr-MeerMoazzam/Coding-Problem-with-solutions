```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'timeConversion' function below.
#
# The function is expected to return a STRING.
# The function accepts STRING s as parameter.
#

def timeConversion(s):
    # Write your code here
    hours=int(s.split(":")[0])
    rem_s=s[2:-2]
    if s[-1:-3:-1][::-1]=="PM":
        if hours<12:
            return str(hours+12)+rem_s
        elif hours==12:
            return str(hours)+rem_s
        return str(0)+str(hours)+rem_s
    else:
        if hours==12:
            return str(0)+str(hours-12)+rem_s
        return str(0)+str(hours)+rem_s
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```
