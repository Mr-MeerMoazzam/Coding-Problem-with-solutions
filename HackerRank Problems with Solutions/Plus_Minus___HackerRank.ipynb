```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'plusMinus' function below.
#
# The function accepts INTEGER_ARRAY arr as parameter.
#

def plusMinus(arr):
    # Write your code here
    n=len(arr)
    zeros=arr.count(0)
    newlist=['>1' if i>0 else '<1' for i in arr]  
    positives=newlist.count('>1')
    negatives=newlist.count('<1')-zeros
    print(f"{positives/n:.6f}")
    print(f"{negatives/n:.6f}")
    print(f"{zeros/n:.6f}")
    
if __name__ == '__main__':
    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    plusMinus(arr)
```
